```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("./data/hospital.csv")
```

# Inspect Data

```python
data.head()
```

```python
# inspect data type and shape

print("df type:", type(data))
print("df shape:", data.shape)
```

```python
# check unique values
data.nunique()
```

```python
# check missing values

data.isnull().sum()
```

```python
# visualize missing values

plt.figure(figsize=(12,6))
sns.heatmap(data.isnull(), cbar=True, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```
